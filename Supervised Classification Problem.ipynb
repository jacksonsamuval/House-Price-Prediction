```python
# Import numerical python 
import numpy as np

# import the pandas
import pandas as pd

# import the seaborn
import seaborn as sns

# import the matplotlib
import matplotlib.pyplot as plt
```

```python
# Load the data
df_House_Price=pd.read_csv('House_Price1.csv')
```

```python
# show the first five observation of the data
df_House_Price.head()
```

```python
# lets perform the data cleaning
df_House_Price.columns
```

```python
dictionary={
    'price':'Price',
    'resid_area':'Residential Area',
    'air_qual':'Air_Quality',
    'room_num':'Room_Number',
    'age':'Age',
    'distance':'Distance',
    'teachers':'Teachers',
    'poor_prop':'Poor_Property',
    'airport':'Airport',
    'n_hos_beds':'N_Hospital_Beds',
    'n_hot_rooms':'N_Hotel_Rooms',
    'waterbody':'Waterbody',
    'rainfall':'Rainfall',
    'bus_ter':'Bus_Terminal',
    'Sold':'Sold'
}
```

```python
df_House_Price=df_House_Price.rename(columns=dictionary)
```

```python
# perform the rename operation
df_House_Price.rename(columns=dictionary)
```

```python
# lets treat the null value in waterbody column
df_House_Price.Waterbody=df_House_Price.Waterbody.fillna('River')
```

```python
# in n_hos_beds, we have less than 10% of data is missing
df_House_Price.dropna(inplace=True)
```

```python
df_target=df_House_Price['Sold']
df_feature=df_House_Price.drop('Sold', axis=1)
```

```python
# segregate the data based on the datatype 
df_num=df_House_Price.select_dtypes(include='number')
df_cat=df_House_Price.select_dtypes(include='object')
```

```python
# use Dummy Encoding
df_encoded=pd.get_dummies(df_cat)

# reset the index
df_encoded=df_encoded.reset_index()
```

```python
# import the standard scalar
from sklearn.preprocessing import StandardScaler

#initiate the standard scalar
ss=StandardScaler()

# scaled the value present in numerical dataframe
scaled=ss.fit_transform(df_num)

# numerate the data into the dataframe
scaled_df=pd.DataFrame(scaled,columns=df_num.columns)

# reset the index
scaled_df=scaled_df.reset_index()
```

```python
scaled
```

```python
 # concat encoded variables and scaled variables
df_independent=pd.concat([df_encoded,scaled_df],axis=1)
```

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data 
x_train,x_test,y_train,y_test=train_test_split(df_independent,
                                              df_target,
                                              test_size=0.2)
```

```python
# import the model
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import classification_report

tree_model = DecisionTreeClassifier()

tree_model.fit(x_train,y_train)

predictions = tree_model.predict(x_test)

class_report=classification_report(y_test,predictions)
print(class_report)
```
